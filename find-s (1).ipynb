```python
def isValue(ele):
    if(ele != "?" and ele != "@"):
        return True
    return False

print("Enter the number of attributes: ")
n = input()
n = int(n)

print("Enter the number of tuples")
t = input()
t = int(t)
attributes = [""]*n
print("Enter all the attributes")
for i in range(n):
    attributes[i] = input()
tuples = []
for i in range(t):
    print("Enter tuple: ")
    values = input()
    list = values.split(",")
    
    tuples.append(list)

h = ["@" for k in range(n-1)]

for tuple in tuples:
    if(tuple[n-1] == "False"):
        continue
    else:
        for j in range(n-1):
            if(h[j] == "@" and isValue(tuple[j])):
                h[j] = tuple[j]
            elif ((isValue(h[j]) and isValue(tuple[j])) and (h[j]!=tuple[j])):
                h[j] = "?"
    
print(h)
```
